```python
import requests

from bs4 import BeautifulSoup
```

```python
url = 'https://lol.fandom.com/wiki/CBLOL/2024_Season/Split_1'
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text, 'html')
```

```python
matchlist = soup.find_all(class_='matchlist-tab-wrapper')
```

```python
tbody = []
for matchweek in matchlist:
  tbody.append(matchweek.find_all('tbody'))
  print(tbody)
```

```python
matches = []

for week in tbody:
  for teste in week:
    matches.append(teste.find_all('tr', class_='ml-row'))
```

```python
single_match = []
for match in matches:
  for single in match:
    single_match.append(single)
```
